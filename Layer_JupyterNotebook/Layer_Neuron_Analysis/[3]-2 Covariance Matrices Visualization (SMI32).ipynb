# Import

## Helper Functions

```python
%store -r analysis_hf
sys.path.insert(0, analysis_hf)
import visualization as viz
```

# Set Directory Paths

```python
# Directory path to load formatted Layer Data
%store -r layer_dataDir

# Directory path to save Calcualted Analysis Data
%store -r layer_CalData

# Directory path to save Figures
%store -r layer_Fig
```

# Loading

## Loading Covariance Matrices

```python
# cov_SMI32_dict
with open(os.path.join(layer_CalData, 'cov_SMI32_dict.pkl'), 'rb') as f:
    cov_SMI32_dict = pickle.load(f)
f.close()
```

## Loading Layer Neuron Labels (SMI32)

```python
# SMI32_Labels
with open(os.path.join(layer_dataDir, 'SMI32_Labels.pkl'), 'rb') as f:
    SMI32_Labels = pickle.load(f)
f.close()
```

# Visualize and save Covariance Matrices (SMI32)

## HC

```python
viz.drawCovMatrix(cov_SMI32_dict['HC'], SMI32_Labels, SMI32_Labels, 'SMI32_HC', outputDir = layer_Fig,
                  annot_fontsize = 6, tick_fontsize = 6, annot_bool = True, save=True)
```

## Tau

```python
viz.drawCovMatrix(cov_SMI32_dict['TAU'], SMI32_Labels, SMI32_Labels, 'SMI32_Tau', outputDir = layer_Fig,
                  annot_fontsize = 6, tick_fontsize = 6, annot_bool = True, save=True)
```

# Node Strength Analysis

```python
def nonZeroDegCorr(DataX1, covMatX1, DataX2, covMatX2, DataX3, covMatX3, DataXLabels,
                   title1, title2, title3, titlefont,
                   x_label, y_label, labelfont,
                   outputDir, outputName, linear_regression = False):
    
    # Copy the Covariance Matrix and set negative values as zero (no need to do this because, negative possible)
    covMatXnz1 = covMatX1.copy()
    covMatXnz2 = covMatX2.copy()
    covMatXnz3 = covMatX3.copy()
    
    # Get sum of covariance values for all regions respective to each region
    # Similar to computing the degree of nodes in a Network
    degX1 = np.sum(covMatXnz1, axis=0, where=~np.isnan(covMatXnz1))
    degX2 = np.sum(covMatXnz2, axis=0, where=~np.isnan(covMatXnz2))
    degX3 = np.sum(covMatXnz3, axis=0, where=~np.isnan(covMatXnz3))

    # Get Y-axis min and max
    combData = np.concatenate((np.nanmean(DataX1, axis=0), 
                               np.nanmean(DataX2, axis=0),
                               np.nanmean(DataX3, axis=0)))

    ymin = np.min(combData) - 0.1
    ymax = np.max(combData) + 0.1
    
    # Define figure
    fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(9, 8))
    
    # Data 1

    axs[0].scatter(degX1, np.nanmean(DataX1, axis=0), facecolors='none', edgecolors='b', linewidths=1, s=30)
    # Get r and p-value
    r1, p1 = scipy.stats.pearsonr(degX1, np.nanmean(DataX1, axis=0))
    
    # Data 2
    axs[1].scatter(degX2, np.nanmean(DataX2, axis=0), facecolors='none', edgecolors='b', linewidths=1, s=30)
    # Get r and p-value
    r2, p2 = scipy.stats.pearsonr(degX2, np.nanmean(DataX2, axis=0))

    # Data 3
    axs[2].scatter(degX3, np.nanmean(DataX3, axis=0), facecolors='none', edgecolors='b', linewidths=1, s=30)
    # Get r and p-value
    r3, p3 = scipy.stats.pearsonr(degX3, np.nanmean(DataX3, axis=0))

    # for i, label in enumerate(DataXLabels):
    #     axs[0].annotate(label, (degX1[i], np.nanmean(DataX1, axis=0)[i]))
    #     axs[1].annotate(label, (degX2[i], np.nanmean(DataX2, axis=0)[i]))
    #     axs[2].annotate(label, (degX3[i], np.nanmean(DataX3, axis=0)[i]))
   
    # set y-axis range
    axs[0].set_ylim([ymin, ymax])
    axs[1].set_ylim([ymin, ymax])
    axs[2].set_ylim([ymin, ymax])

    # Set title
    #plt.title(title + f", r={r:.6f}, p={p:.6f}", fontsize=10)
    axs[0].set_title(title1, fontsize = titlefont)
    axs[1].set_title(title2, fontsize = titlefont)
    axs[2].set_title(title3, fontsize = titlefont)
    
    # Draw Linear Regression Line (is set to True)
    if linear_regression:
        # Obtain m (slope) and b(intercept) of linear regression line
        m1, b1 = np.polyfit(degX1, np.nanmean(DataX1, axis=0), 1)
        m2, b2 = np.polyfit(degX2, np.nanmean(DataX2, axis=0), 1)
        m3, b3 = np.polyfit(degX3, np.nanmean(DataX3, axis=0), 1)
        
        #add linear regression line to scatterplot 
        axs[0].plot(degX1, m1*degX1+b1, color="red")
        axs[1].plot(degX2, m2*degX2+b2, color="red")
        axs[2].plot(degX3, m3*degX3+b3, color="red")

    # Set X and Y Labels
    axs[0].set_xlabel(x_label, fontsize = labelfont)
    axs[0].set_ylabel(y_label, fontsize = labelfont)
    axs[1].set_xlabel(x_label, fontsize = labelfont)
    axs[2].set_xlabel(x_label, fontsize = labelfont)
    
    # Record r and p values
    if p1 < 0.05:
        axs[0].text(0.5, -0.12, f'r={r1: .4f}, *p={p1: .4f}', ha='center', va='center', transform=axs[0].transAxes, fontsize=12)
    else:
        axs[0].text(0.5, -0.12, f'r={r1: .4f}, p={p1: .4f}', ha='center', va='center', transform=axs[0].transAxes, fontsize=12)
    
    if p2 < 0.05:
        axs[1].text(0.5, -0.12, f'r={r2: .4f}, *p={p2: .4f}', ha='center', va='center', transform=axs[1].transAxes, fontsize=12)
    else:
        axs[1].text(0.5, -0.12, f'r={r2: .4f}, p={p2: .4f}', ha='center', va='center', transform=axs[1].transAxes, fontsize=12)

    if p3 < 0.05:
        axs[2].text(0.5, -0.12, f'r={r3: .4f}, *p={p3: .4f}', ha='center', va='center', transform=axs[2].transAxes, fontsize=12)
    else:
        axs[2].text(0.5, -0.12, f'r={r3: .4f}, p={p3: .4f}', ha='center', va='center', transform=axs[2].transAxes, fontsize=12)
    
    # Show Figure
    plt.show()

    # # Save Figure
    # plt.savefig(os.path.join(outputDir, outputName) + '.png', dpi=400)
    
    
```

# Divide into Long vs Short Graphs (using covariance matrix)

## Helper Function to make a upper triangle matrix symmetric

```python
def symmetric_uppertri(covMat):
    # Convert NaNs to zeros
    upper_triangular_matrix_no_nan = np.nan_to_num(covMat)

    # Make the matrix symmetric by mirroring the upper triangular part to the lower triangular part
    symmetric_matrix = upper_triangular_matrix_no_nan + upper_triangular_matrix_no_nan.T - np.diag(upper_triangular_matrix_no_nan.diagonal())
    symmetric_matrix = np.where(symmetric_matrix == 0, np.nan, symmetric_matrix)

    return symmetric_matrix
```

## Helper Function to draw Node Strength correlation plot for Long & Short paths

```python
def corrPlot_longshort(curr_data, curr_cov, neuron_type, group_type):
    # Define long & short cov matrix
    curr_cov_long = np.full(curr_cov.shape, np.nan)
    curr_cov_short = np.full(curr_cov.shape, np.nan)

    # Fill the long & short cov matrix
    for i in range(curr_cov.shape[0]):
        for j in range(curr_cov.shape[1]): 
            if i < j:# we want i < j
                # Get the source and target
                src = Labels[i]
                tgt = Labels[j]

                if cp.condition(src, tgt, 'canonical'): # Fill the long covMat
                    curr_cov_long[i][j] = curr_cov[i][j]

                if cp.condition(src, tgt, 'short'): # Fill the short covMat
                    curr_cov_short[i][j] = curr_cov[i][j]


    # make the long & short cov matrix symmetric
    curr_cov_long = symmetric_uppertri(curr_cov_long)
    curr_cov_short = symmetric_uppertri(curr_cov_short)
    
    # Find indices of columns that contain at least one non-NaN element
    column_indices_long = np.where(~np.isnan(curr_cov_long).all(axis=0))[0] 
    column_indices_short = np.where(~np.isnan(curr_cov_short).all(axis=0))[0] 

    # Modify the long, short cov matrix
    curr_cov_long = curr_cov_long[:, column_indices_long]
    curr_cov_short = curr_cov_short[:, column_indices_short]

    # Modify the curr_data
    curr_data_long = curr_data[:, column_indices_long]
    curr_data_short = curr_data[:, column_indices_short]

    # Plot the correlation plot
    nonZeroDegCorr(curr_data, curr_cov, curr_data_long, curr_cov_long, curr_data_short, curr_cov_short, Labels,
                f'{neuron_type}-{group_type}(Whole)', f'{neuron_type}-{group_type}(Long)', f'{neuron_type}-{group_type}(Short)', 
                14,'Node Strength', 'Mean %AO', 15,
                layer_Fig, 'FTD_TauTDP_NodeStrenght(paperfigure1)', 
                linear_regression = True)
```

```python
Labels = ['L23_1(Cingulate)', 'L23_2(Cingulate)', 'L23_3d(Paracingulate)', 'L23_3v(Paracingulate)', 'L23_4(Rectus)', 'L23_5(Middle Frontal)', 
               'L56_1(Cingulate)', 'L56_2(Cingulate)', 'L56_3d(Paracingulate)', 'L56_3v(Paracingulate)', 'L56_4(Rectus)', 'L56_5(Middle Frontal)']
```

```python
sexHC_L = np.array(NeuN_HC_Sex_L).reshape(-1, 1)
sexTau_L = np.array(NeuN_Tau_Sex_L).reshape(-1, 1)
sexTDP_L = np.array(NeuN_TDP_Sex_L).reshape(-1, 1)

sexHC_R = np.array(NeuN_HC_Sex_R).reshape(-1, 1)
sexTau_R = np.array(NeuN_Tau_Sex_R).reshape(-1, 1)
sexTDP_R = np.array(NeuN_TDP_Sex_R).reshape(-1, 1)

# Convert Male --> 0 / Female --> 1
sexHC_L = np.array(sb.sexToBinary(sexHC_L)).reshape((-1, 1)) 
sexTau_L = np.array(sb.sexToBinary(sexTau_L)).reshape((-1, 1)) 
sexTDP_L = np.array(sb.sexToBinary(sexTDP_L)).reshape((-1, 1)) 

sexHC_R = np.array(sb.sexToBinary(sexHC_R)).reshape((-1, 1)) 
sexTau_R = np.array(sb.sexToBinary(sexTau_R)).reshape((-1, 1)) 
sexTDP_R = np.array(sb.sexToBinary(sexTDP_R)).reshape((-1, 1)) 
```

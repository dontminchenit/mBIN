# Import

```python
import sys
import os
import pickle
```

## Helper Functions

```python
%store -r analysis_hf
sys.path.insert(0, analysis_hf)
import visualization as viz
```

# Set Directory Paths

```python
# Directory path to load formatted Layer Data
%store -r layer_dataDir

# Directory path to save Calcualted Analysis Data
%store -r layer_CalData

# Directory path to save Figures
%store -r layer_Fig
```

# Loading

## Loading Covariance Matrices

```python
# cov_NeuN_dict
with open(os.path.join(layer_CalData, 'cov_NeuN_dict.pkl'), 'rb') as f:
    cov_NeuN_dict = pickle.load(f)
f.close()
```

## Loading Layer Neuron Labels (NeuN)

```python
# NeuN_Labels
with open(os.path.join(layer_dataDir, 'NeuN_Labels.pkl'), 'rb') as f:
    NeuN_Labels = pickle.load(f)
f.close()
```

# Visualize and save Covariance Matrices (NeuN)

## HC

```python
viz.drawCovMatrix(cov_NeuN_dict['HC'], NeuN_Labels, NeuN_Labels, 'NeuN_HC', outputDir = layer_Fig,
                  annot_fontsize = 6, tick_fontsize = 6, annot_bool = True, save=True)
```

## Tau

```python
viz.drawCovMatrix(cov_NeuN_dict['TAU'], NeuN_Labels, NeuN_Labels, 'NeuN_Tau', outputDir = layer_Fig,
                  annot_fontsize = 6, tick_fontsize = 6, annot_bool = True, save=True)
```

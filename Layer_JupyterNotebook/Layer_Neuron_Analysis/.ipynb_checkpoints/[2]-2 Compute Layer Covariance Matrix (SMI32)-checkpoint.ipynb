## TDP

```python
# SMI32_TDP
with open(os.path.join(layer_dataDir, 'SMI32_TDP.pkl'), 'rb') as f:
    SMI32_TDP = pickle.load(f)
f.close()

# SMI32_TDP_w
with open(os.path.join(layer_dataDir, 'SMI32_TDP_w.pkl'), 'rb') as f:
    SMI32_TDP_w = pickle.load(f)
f.close()
```

# Set Parameters

## p-value threshold

```python
# pthresh_list = [0.05]
pthresh = 0.05
```

## Covariance Matrix Threshold (to remove noise)

```python
cov_thresh = 0.1
```

# Calculate Covariance Matrices (SMI32)

## HC, Tau, TDP, Tau < HC, Tau > HC, TDP < HC, TDP > HC, Tau > TDP, TDP > Tau

```python
cov_SMI32_dict = c.calCovMatDict(SMI32_HC, SMI32_Tau, SMI32_TDP, pthresh, cov_thresh)
```

```python
cov_SMI32_dict_w = c.calCovMatDict(SMI32_HC_w, SMI32_Tau_w, SMI32_TDP_w, pthresh, cov_thresh)
```

# Saving

```python
# cov_SMI32_dict
with open(os.path.join(layer_CalData, 'cov_SMI32_dict.pkl'), 'wb') as f:
    pickle.dump(cov_SMI32_dict, f)
f.close()
```

```python
# cov_SMI32_dict_w
with open(os.path.join(layer_CalData, 'cov_SMI32_dict_w.pkl'), 'wb') as f:
    pickle.dump(cov_SMI32_dict_w, f)
f.close()
```

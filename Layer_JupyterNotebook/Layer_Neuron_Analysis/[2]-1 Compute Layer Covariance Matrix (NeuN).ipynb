# Set Parameters

## p-value threshold

```python
# pthresh_list = [0.05]
pthresh = 0.05
```

## Covariance Matrix Threshold (to remove noise)

```python
cov_thresh = -1.5
```

# Calculate Covariance Matrices (NeuN)

## HC, Tau, TDP, Tau < HC, Tau > HC, TDP < HC, TDP > HC, Tau > TDP, TDP > Tau

```python
cov_NeuN_dict = c.calCovMatDict(NeuN_HC, NeuN_Tau, NeuN_TDP, pthresh, cov_thresh)
```

```python
cov_NeuN_dict_w_agesex = c.calCovMatDict(NeuN_HC_w_agesex, NeuN_Tau_w_agesex, NeuN_TDP_w_agesex, pthresh, cov_thresh)
```

```python
cov_NeuN_dict_w_agesexNeuNADNC = c.calCovMatDict(NeuN_HC_w_agesexNeuNADNC, NeuN_Tau_w_agesexNeuNADNC, NeuN_TDP_w_agesexNeuNADNC, pthresh, cov_thresh)
```

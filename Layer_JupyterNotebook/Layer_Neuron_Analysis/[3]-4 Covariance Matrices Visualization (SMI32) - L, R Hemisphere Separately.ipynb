# Set Directory Paths

```python
# Directory path to load formatted Layer Data
%store -r layer_dataDir

# Directory path to save Calcualted Analysis Data
%store -r layer_CalData

# Directory path to save Figures
%store -r layer_Fig
```

# Loading

## Loading Covariance Matrices

```python
# cov_SMI32_dict_L
with open(os.path.join(layer_CalData, 'cov_SMI32_dict_L.pkl'), 'rb') as f:
    cov_SMI32_dict_L = pickle.load(f)
f.close()

# cov_SMI32_dict_R
with open(os.path.join(layer_CalData, 'cov_SMI32_dict_R.pkl'), 'rb') as f:
    cov_SMI32_dict_R = pickle.load(f)
f.close()
```

```python
# cov_SMI32_dict_w_L
with open(os.path.join(layer_CalData, 'cov_SMI32_dict_w_L.pkl'), 'rb') as f:
    cov_SMI32_dict_w_L = pickle.load(f)
f.close()

# cov_SMI32_dict_w_R
with open(os.path.join(layer_CalData, 'cov_SMI32_dict_w_R.pkl'), 'rb') as f:
    cov_SMI32_dict_w_R = pickle.load(f)
f.close()
```

## Loading Layer Neuron Labels (SMI32)

```python
# SMI32_Labels
with open(os.path.join(layer_dataDir, 'SMI32_Labels.pkl'), 'rb') as f:
    SMI32_Labels = pickle.load(f)
f.close()
```

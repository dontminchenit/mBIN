# Calculating W-Score

## Format Age Values

```python
ageHC = np.array(SMI32_HC_Age).reshape(-1, 1)
ageTau = np.array(SMI32_Tau_Age).reshape(-1, 1)
ageTDP = np.array(SMI32_TDP_Age).reshape(-1, 1)
```

## Format Sex Values

```python
sexHC = np.array(SMI32_HC_Sex).reshape(-1, 1)
sexTau = np.array(SMI32_Tau_Sex).reshape(-1, 1)
sexTDP = np.array(SMI32_TDP_Sex).reshape(-1, 1)


# Convert Male --> 0 / Female --> 1
sexHC = np.array(sb.sexToBinary(sexHC)).reshape((-1, 1)) 
sexTau = np.array(sb.sexToBinary(sexTau)).reshape((-1, 1)) 
sexTDP = np.array(sb.sexToBinary(sexTDP)).reshape((-1, 1)) 
```

## Format L1-6 values

```python
SMI32_HC_NeuN_1 = np.array(SMI32_HC_NeuN_1).reshape(-1, 1)
SMI32_HC_NeuN_2 = np.array(SMI32_HC_NeuN_2).reshape(-1, 1)
SMI32_HC_NeuN_3d = np.array(SMI32_HC_NeuN_3d).reshape(-1, 1)
SMI32_HC_NeuN_3v = np.array(SMI32_HC_NeuN_3v).reshape(-1, 1)
SMI32_HC_NeuN_4 = np.array(SMI32_HC_NeuN_4).reshape(-1, 1)
SMI32_HC_NeuN_5 = np.array(SMI32_HC_NeuN_5).reshape(-1, 1)

SMI32_Tau_NeuN_1 = np.array(SMI32_Tau_NeuN_1).reshape(-1, 1)
SMI32_Tau_NeuN_2 = np.array(SMI32_Tau_NeuN_2).reshape(-1, 1)
SMI32_Tau_NeuN_3d = np.array(SMI32_Tau_NeuN_3d).reshape(-1, 1)
SMI32_Tau_NeuN_3v = np.array(SMI32_Tau_NeuN_3v).reshape(-1, 1)
SMI32_Tau_NeuN_4 = np.array(SMI32_Tau_NeuN_4).reshape(-1, 1)
SMI32_Tau_NeuN_5 = np.array(SMI32_Tau_NeuN_5).reshape(-1, 1)

SMI32_TDP_NeuN_1 = np.array(SMI32_TDP_NeuN_1).reshape(-1, 1)
SMI32_TDP_NeuN_2 = np.array(SMI32_TDP_NeuN_2).reshape(-1, 1)
SMI32_TDP_NeuN_3d = np.array(SMI32_TDP_NeuN_3d).reshape(-1, 1)
SMI32_TDP_NeuN_3v = np.array(SMI32_TDP_NeuN_3v).reshape(-1, 1)
SMI32_TDP_NeuN_4 = np.array(SMI32_TDP_NeuN_4).reshape(-1, 1)
SMI32_TDP_NeuN_5 = np.array(SMI32_TDP_NeuN_5).reshape(-1, 1)
```

## Format ADNC values

```python
ADNCHC = np.array(SMI32_HC_ADNC).reshape(-1, 1)
ADNCTau = np.array(SMI32_Tau_ADNC).reshape(-1, 1)
ADNCTDP = np.array(SMI32_TDP_ADNC).reshape(-1, 1)

# Convert Male --> 0 / Female --> 1
ADNCHC = np.array(sb.ADNCToBinary(ADNCHC)).reshape((-1, 1)) 
ADNCTau = np.array(sb.ADNCToBinary(ADNCTau)).reshape((-1, 1)) 
ADNCTDP = np.array(sb.ADNCToBinary(ADNCTDP)).reshape((-1, 1)) 
```

## Stack Age, Sex Values

```python
# Stack the Age and Sex features into 2D array
hcCov_CT1 = np.concatenate((ageHC, sexHC, SMI32_HC_NeuN_1, ADNCHC), axis=1) 
hcCov_CT2 = np.concatenate((ageHC, sexHC, SMI32_HC_NeuN_2, ADNCHC), axis=1) 
hcCov_CT3d = np.concatenate((ageHC, sexHC, SMI32_HC_NeuN_3d, ADNCHC), axis=1) 
hcCov_CT3v = np.concatenate((ageHC, sexHC, SMI32_HC_NeuN_3v, ADNCHC), axis=1) 
hcCov_CT4 = np.concatenate((ageHC, sexHC, SMI32_HC_NeuN_4, ADNCHC), axis=1) 
hcCov_CT5 = np.concatenate((ageHC, sexHC, SMI32_HC_NeuN_5, ADNCHC), axis=1) 

tauCov_CT1 = np.concatenate((ageTau, sexTau, SMI32_Tau_NeuN_1, ADNCTau), axis=1)  
tauCov_CT2 = np.concatenate((ageTau, sexTau, SMI32_Tau_NeuN_2, ADNCTau), axis=1)  
tauCov_CT3d = np.concatenate((ageTau, sexTau, SMI32_Tau_NeuN_3d, ADNCTau), axis=1) 
tauCov_CT3v = np.concatenate((ageTau, sexTau, SMI32_Tau_NeuN_3v, ADNCTau), axis=1)  
tauCov_CT4 = np.concatenate((ageTau, sexTau, SMI32_Tau_NeuN_4, ADNCTau), axis=1)  
tauCov_CT5 = np.concatenate((ageTau, sexTau, SMI32_Tau_NeuN_5, ADNCTau), axis=1)  

tdpCov_CT1 = np.concatenate((ageTDP, sexTDP, SMI32_TDP_NeuN_1, ADNCTDP), axis=1)
tdpCov_CT2 = np.concatenate((ageTDP, sexTDP, SMI32_TDP_NeuN_2, ADNCTDP), axis=1)
tdpCov_CT3d = np.concatenate((ageTDP, sexTDP, SMI32_TDP_NeuN_3d, ADNCTDP), axis=1)
tdpCov_CT3v = np.concatenate((ageTDP, sexTDP, SMI32_TDP_NeuN_3v, ADNCTDP), axis=1)
tdpCov_CT4 = np.concatenate((ageTDP, sexTDP, SMI32_TDP_NeuN_4, ADNCTDP), axis=1)
tdpCov_CT5 = np.concatenate((ageTDP, sexTDP, SMI32_TDP_NeuN_5, ADNCTDP), axis=1)


# # Stack the Age and Sex features into 2D array
# hcCov_CT1 = np.concatenate((ageHC, sexHC), axis=1) 
# hcCov_CT2 = np.concatenate((ageHC, sexHC), axis=1) 
# hcCov_CT3d = np.concatenate((ageHC, sexHC), axis=1) 
# hcCov_CT3v = np.concatenate((ageHC, sexHC), axis=1) 
# hcCov_CT4 = np.concatenate((ageHC, sexHC), axis=1) 
# hcCov_CT5 = np.concatenate((ageHC, sexHC), axis=1) 

# tauCov_CT1 = np.concatenate((ageTau, sexTau), axis=1)  
# tauCov_CT2 = np.concatenate((ageTau, sexTau), axis=1)  
# tauCov_CT3d = np.concatenate((ageTau, sexTau), axis=1) 
# tauCov_CT3v = np.concatenate((ageTau, sexTau), axis=1)  
# tauCov_CT4 = np.concatenate((ageTau, sexTau), axis=1)  
# tauCov_CT5 = np.concatenate((ageTau, sexTau), axis=1)  

# tdpCov_CT1 = np.concatenate((ageTDP, sexTDP), axis=1)
# tdpCov_CT2 = np.concatenate((ageTDP, sexTDP), axis=1)
# tdpCov_CT3d = np.concatenate((ageTDP, sexTDP), axis=1)
# tdpCov_CT3v = np.concatenate((ageTDP, sexTDP), axis=1)
# tdpCov_CT4 = np.concatenate((ageTDP, sexTDP), axis=1)
# tdpCov_CT5 = np.concatenate((ageTDP, sexTDP), axis=1)
```

## Divide HC, Tau, TDP into Cortical Types

```python
SMI32_HC_CT1 = SMI32_HC[:, [0, 0+6]]
SMI32_HC_CT2 = SMI32_HC[:, [1, 1+6]]
SMI32_HC_CT3d = SMI32_HC[:, [2, 2+6]]
SMI32_HC_CT3v = SMI32_HC[:, [3, 3+6]]
SMI32_HC_CT4 = SMI32_HC[:, [4, 4+6]]
SMI32_HC_CT5 = SMI32_HC[:, [5, 5+6]]

SMI32_Tau_CT1 = SMI32_Tau[:, [0, 0+6]]
SMI32_Tau_CT2 = SMI32_Tau[:, [1, 1+6]]
SMI32_Tau_CT3d = SMI32_Tau[:, [2, 2+6]]
SMI32_Tau_CT3v = SMI32_Tau[:, [3, 3+6]]
SMI32_Tau_CT4 = SMI32_Tau[:, [4, 4+6]]
SMI32_Tau_CT5 = SMI32_Tau[:, [5, 5+6]]

SMI32_TDP_CT1 = SMI32_TDP[:, [0, 0+6]]
SMI32_TDP_CT2 = SMI32_TDP[:, [1, 1+6]]
SMI32_TDP_CT3d = SMI32_TDP[:, [2, 2+6]]
SMI32_TDP_CT3v = SMI32_TDP[:, [3, 3+6]]
SMI32_TDP_CT4 = SMI32_TDP[:, [4, 4+6]]
SMI32_TDP_CT5 = SMI32_TDP[:, [5, 5+6]]
```

## Number of regions we are analyzing (N = 10)

```python
#N = SMI32_HC_L.shape[1]
N=2
```

## Calculate the W-Scores

```python
# L
## CT1
SMI32_HC_w_CT1, SMI32_Tau_w_CT1, SMI32_TDP_w_CT1 = wScore.generateWScore(hcCov_CT1, tauCov_CT1, tdpCov_CT1, 
                                                                            N, 
                                                                            SMI32_HC_CT1, 
                                                                            SMI32_Tau_CT1,  
                                                                            SMI32_TDP_CT1)
## CT2
SMI32_HC_w_CT2, SMI32_Tau_w_CT2, SMI32_TDP_w_CT2 = wScore.generateWScore(hcCov_CT2, tauCov_CT2, tdpCov_CT2, 
                                                                            N, 
                                                                            SMI32_HC_CT2, 
                                                                            SMI32_Tau_CT2,  
                                                                            SMI32_TDP_CT2)
## CT3d
SMI32_HC_w_CT3d, SMI32_Tau_w_CT3d, SMI32_TDP_w_CT3d = wScore.generateWScore(hcCov_CT3d, tauCov_CT3d, tdpCov_CT3d, 
                                                                            N, 
                                                                            SMI32_HC_CT3d, 
                                                                            SMI32_Tau_CT3d,  
                                                                            SMI32_TDP_CT3d)

## CT3v
SMI32_HC_w_CT3v, SMI32_Tau_w_CT3v, SMI32_TDP_w_CT3v = wScore.generateWScore(hcCov_CT3v, tauCov_CT3v, tdpCov_CT3v, 
                                                                            N, 
                                                                            SMI32_HC_CT3v, 
                                                                            SMI32_Tau_CT3v,  
                                                                            SMI32_TDP_CT3v)
## CT4
SMI32_HC_w_CT4, SMI32_Tau_w_CT4, SMI32_TDP_w_CT4 = wScore.generateWScore(hcCov_CT4, tauCov_CT4, tdpCov_CT4, 
                                                                            N, 
                                                                            SMI32_HC_CT4, 
                                                                            SMI32_Tau_CT4,  
                                                                            SMI32_TDP_CT4)
## CT5
SMI32_HC_w_CT5, SMI32_Tau_w_CT5, SMI32_TDP_w_CT5 = wScore.generateWScore(hcCov_CT5, tauCov_CT5, tdpCov_CT5, 
                                                                            N, 
                                                                            SMI32_HC_CT5, 
                                                                            SMI32_Tau_CT5,  
                                                                            SMI32_TDP_CT5)
```

```python
SMI32_HC_w = np.zeros(SMI32_HC.shape)
SMI32_Tau_w = np.zeros(SMI32_Tau.shape)
SMI32_TDP_w = np.zeros(SMI32_TDP.shape)

for i in range(6):
    if i == 0:
        SMI32_HC_w[:, [i, i+6]] = SMI32_HC_w_CT1
        SMI32_Tau_w[:, [i, i+6]] = SMI32_Tau_w_CT1
        SMI32_TDP_w[:, [i, i+6]] = SMI32_TDP_w_CT1
        
    elif i == 1:
        SMI32_HC_w[:, [i, i+6]] = SMI32_HC_w_CT2
        SMI32_Tau_w[:, [i, i+6]] = SMI32_Tau_w_CT2
        SMI32_TDP_w[:, [i, i+6]] = SMI32_TDP_w_CT2
        
    elif i == 2:
        SMI32_HC_w[:, [i, i+6]] = SMI32_HC_w_CT3d
        SMI32_Tau_w[:, [i, i+6]] = SMI32_Tau_w_CT3d
        SMI32_TDP_w[:, [i, i+6]] = SMI32_TDP_w_CT3d

    elif i == 3:
        SMI32_HC_w[:, [i, i+6]] = SMI32_HC_w_CT3v
        SMI32_Tau_w[:, [i, i+6]] = SMI32_Tau_w_CT3v
        SMI32_TDP_w[:, [i, i+6]] = SMI32_TDP_w_CT3v
        
    elif i == 4:
        SMI32_HC_w[:, [i, i+6]] = SMI32_HC_w_CT4
        SMI32_Tau_w[:, [i, i+6]] = SMI32_Tau_w_CT4
        SMI32_TDP_w[:, [i, i+6]] = SMI32_TDP_w_CT4
        
    else:
        SMI32_HC_w[:, [i, i+6]] = SMI32_HC_w_CT5
        SMI32_Tau_w[:, [i, i+6]] = SMI32_Tau_w_CT5
        SMI32_TDP_w[:, [i, i+6]] = SMI32_TDP_w_CT5
```

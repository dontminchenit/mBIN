# Loading

## Loading Normalized Volume at Pathology Regions / TAU, TDP --> Dropped

```python
# HCnormVolumeAtPath
with open(os.path.join(thickAtPath_dataDir, 'HCnormVolumeAtPath.pkl'), 'rb') as f:
    HCnormVolumeAtPath = pickle.load(f)
f.close()

# TAUnormVolumeAtPath_Drop
with open(os.path.join(thickAtPath_dataDir, 'TAUnormVolumeAtPath_Drop.pkl'), 'rb') as f:
    TAUnormVolumeAtPath_Drop = pickle.load(f)
f.close()

# TDPnormVolumeAtPath_Drop
with open(os.path.join(thickAtPath_dataDir, 'TDPnormVolumeAtPath_Drop.pkl'), 'rb') as f:
    TDPnormVolumeAtPath_Drop = pickle.load(f)
f.close()
```

## Loading Normalized Volume at Pathology Regions - Z Score / TAU, TDP --> Dropped

```python
# HCnormVolAtPath_z
with open(os.path.join(thickAtPath_CalData, 'HCnormVolAtPath_z.pkl'), 'rb') as f:
    HCnormVolAtPath_z = pickle.load(f)
f.close()

# TAUnormVolAtPath_z_Drop
with open(os.path.join(thickAtPath_CalData, 'TAUnormVolAtPath_z_Drop.pkl'), 'rb') as f:
    TAUnormVolAtPath_z_Drop = pickle.load(f)
f.close()

# TDPnormVolAtPath_z_Drop
with open(os.path.join(thickAtPath_CalData, 'TDPnormVolAtPath_z_Drop.pkl'), 'rb') as f:
    TDPnormVolAtPath_z_Drop = pickle.load(f)
f.close()
```

## Loading Normalized Volume at Pathology Regions - W Score / TAU, TDP --> Dropped

```python
# HCnormVolAtPath_w
with open(os.path.join(thickAtPath_CalData, 'HCnormVolAtPath_w.pkl'), 'rb') as f:
    HCnormVolAtPath_w = pickle.load(f)
f.close()

# TAUnormVolAtPath_w_Drop
with open(os.path.join(thickAtPath_CalData, 'TAUnormVolAtPath_w_Drop.pkl'), 'rb') as f:
    TAUnormVolAtPath_w_Drop = pickle.load(f)
f.close()

# TDPnormVolAtPath_w_Drop
with open(os.path.join(thickAtPath_CalData, 'TDPnormVolAtPath_w_Drop.pkl'), 'rb') as f:
    TDPnormVolAtPath_w_Drop = pickle.load(f)
f.close()
```

# Generate Box-plopt of Normalized Volume At Pathlogy - Original/Z Score/W Score Distributions (Mean across subjects)

## Distribution of Normalized Volume At Pathology - Original

```python
bp.drawthicknessboxplot(HCnormVolumeAtPath, TAUnormVolumeAtPath_Drop, TDPnormVolumeAtPath_Drop, ['HC', 'TAU', 'TDP'], 
                     'Mean Normalized Volume', "NormVolAtPath Original Value Distribution", thickAtPath_Fig, 
                        "FTD_Distribution_NormVolAtPath_Original.png")
```

## Distribution of Normalized Volume At Pathology - Z Score

```python
bp.drawthicknessboxplot(HCnormVolAtPath_z, TAUnormVolAtPath_z_Drop, TDPnormVolAtPath_z_Drop, ['HC', 'TAU', 'TDP'], 
                     'Mean Normalized Volume', "NormVolAtPath Z Score Distribution", thickAtPath_Fig, 
                        "FTD_Distribution_NormVolAtPath_ZScore.png")
```

## Distribution of Normalized Volume At Pathology - W Score

```python
bp.drawthicknessboxplot(HCnormVolAtPath_w, TAUnormVolAtPath_w_Drop, TDPnormVolAtPath_w_Drop, ['HC', 'TAU', 'TDP'], 
                     'Mean Normalized Volume', "NormVolAtPath W Score Distribution", thickAtPath_Fig, 
                        "FTD_Distribution_NormVolAtPath_WScore.png")
```

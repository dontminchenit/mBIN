# Draw Nodal Strength vs Thickness Values

## Thickness Data - Original

### HC

```python
ns.nonZeroDegCorr(thickHC, cov_thick_dict["HC"], 'FTD_Thickness_HC', 'Degree', 'Mean Thickness', 
                  outputDir = thick_Fig, outputName = 'FTD_NodalStrength_Thickness_Orig_(HC)', 
                  linear_regression = True)
```

### TAU

```python
ns.nonZeroDegCorr(thickTAU, cov_thick_dict["TAU"], 'FTD_Thickness_TAU', 'Degree', 'Mean Thickness', 
                  outputDir = thick_Fig, outputName = 'FTD_NodalStrength_Thickness_Orig_(TAU)', 
                  linear_regression = True)
```

### TDP

```python
ns.nonZeroDegCorr(thickTDP, cov_thick_dict["TDP"], 'FTD_Thickness_TDP', 'Degree', 'Mean Thickness', 
                  outputDir = thick_Fig, outputName = 'FTD_NodalStrength_Thickness_Orig_(TDP)', 
                  linear_regression = True)
```

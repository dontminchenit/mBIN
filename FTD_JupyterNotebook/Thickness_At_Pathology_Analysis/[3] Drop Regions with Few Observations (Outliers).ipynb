## Covariance Matrices

### Helper Function

```python
def covMatAtPathDrop(covDict, TAU_missing_index, TDP_missing_index):
    covDict_Drop = covDict.copy()
    
    covDict_Drop["TAU"] = np.delete(covDict_Drop["TAU"], TAU_missing_index, axis = 0)
    covDict_Drop["TDP"] = np.delete(covDict_Drop["TDP"], TDP_missing_index, axis = 0)
    covDict_Drop["TAU_gt_TDP"] = np.delete(covDict_Drop["TAU_gt_TDP"], TAU_missing_index, axis = 0)
    covDict_Drop["TDP_gt_TAU"] = np.delete(covDict_Drop["TDP_gt_TAU"], TDP_missing_index, axis = 0)
    covDict_Drop["TAU_gt_HC"] = np.delete(covDict_Drop["TAU_gt_HC"], TAU_missing_index, axis = 0)
    covDict_Drop["TAU_lt_HC"] = np.delete(covDict_Drop["TAU_lt_HC"], TAU_missing_index, axis = 0)
    covDict_Drop["TDP_gt_HC"] = np.delete(covDict_Drop["TDP_gt_HC"], TDP_missing_index, axis = 0)
    covDict_Drop["TDP_lt_HC"] = np.delete(covDict_Drop["TDP_lt_HC"], TDP_missing_index, axis = 0)
    covDict_Drop["TAU_gt_TDP_raw"] = np.delete(covDict_Drop["TAU_gt_TDP_raw"], TAU_missing_index, axis = 0)
    covDict_Drop["TDP_gt_TAU_raw"] = np.delete(covDict_Drop["TDP_gt_TAU_raw"], TDP_missing_index, axis = 0)
    covDict_Drop["TAU_gt_HC_raw"] = np.delete(covDict_Drop["TAU_gt_HC_raw"], TAU_missing_index, axis = 0)
    covDict_Drop["TAU_lt_HC_raw"] = np.delete(covDict_Drop["TAU_lt_HC_raw"], TAU_missing_index, axis = 0)
    covDict_Drop["TDP_gt_HC_raw"] = np.delete(covDict_Drop["TDP_gt_HC_raw"], TDP_missing_index, axis = 0)
    covDict_Drop["TDP_lt_HC_raw"] = np.delete(covDict_Drop["TDP_lt_HC_raw"], TDP_missing_index, axis = 0)
    
    covDict_Drop["TAU"] = np.delete(covDict_Drop["TAU"], TAU_missing_index, axis = 1)
    covDict_Drop["TDP"] = np.delete(covDict_Drop["TDP"], TDP_missing_index, axis = 1)
    covDict_Drop["TAU_gt_TDP"] = np.delete(covDict_Drop["TAU_gt_TDP"], TAU_missing_index, axis = 1)
    covDict_Drop["TDP_gt_TAU"] = np.delete(covDict_Drop["TDP_gt_TAU"], TDP_missing_index, axis = 1)
    covDict_Drop["TAU_gt_HC"] = np.delete(covDict_Drop["TAU_gt_HC"], TAU_missing_index, axis = 1)
    covDict_Drop["TAU_lt_HC"] = np.delete(covDict_Drop["TAU_lt_HC"], TAU_missing_index, axis = 1)
    covDict_Drop["TDP_gt_HC"] = np.delete(covDict_Drop["TDP_gt_HC"], TDP_missing_index, axis = 1)
    covDict_Drop["TDP_lt_HC"] = np.delete(covDict_Drop["TDP_lt_HC"], TDP_missing_index, axis = 1)
    covDict_Drop["TAU_gt_TDP_raw"] = np.delete(covDict_Drop["TAU_gt_TDP_raw"], TAU_missing_index, axis = 1)
    covDict_Drop["TDP_gt_TAU_raw"] = np.delete(covDict_Drop["TDP_gt_TAU_raw"], TDP_missing_index, axis = 1)
    covDict_Drop["TAU_gt_HC_raw"] = np.delete(covDict_Drop["TAU_gt_HC_raw"], TAU_missing_index, axis = 1)
    covDict_Drop["TAU_lt_HC_raw"] = np.delete(covDict_Drop["TAU_lt_HC_raw"], TAU_missing_index, axis = 1)
    covDict_Drop["TDP_gt_HC_raw"] = np.delete(covDict_Drop["TDP_gt_HC_raw"], TDP_missing_index, axis = 1)
    covDict_Drop["TDP_lt_HC_raw"] = np.delete(covDict_Drop["TDP_lt_HC_raw"], TDP_missing_index, axis = 1)
    
    return covDict_Drop
```

```python
cov_thickAtPath_dict_Drop = covMatAtPathDrop(cov_thickAtPath_dict, TAU_missing_index, TDP_missing_index)
cov_thickAtPath_z_dict_Drop = covMatAtPathDrop(cov_thickAtPath_z_dict, TAU_missing_index, TDP_missing_index)
cov_thickAtPath_w_dict_Drop = covMatAtPathDrop(cov_thickAtPath_w_dict, TAU_missing_index, TDP_missing_index)
```

```python
cov_volAtPath_dict_Drop = covMatAtPathDrop(cov_volAtPath_dict, TAU_missing_index, TDP_missing_index)
cov_volAtPath_z_dict_Drop = covMatAtPathDrop(cov_volAtPath_z_dict, TAU_missing_index, TDP_missing_index)
cov_volAtPath_w_dict_Drop = covMatAtPathDrop(cov_volAtPath_w_dict, TAU_missing_index, TDP_missing_index)
```
